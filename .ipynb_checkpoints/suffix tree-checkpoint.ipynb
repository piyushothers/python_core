```python
class SuffixTree():
    def __init__(self,word):
        self.word = word
        self.charToFirstWord = dict()
        self.nodeCount = 0
        root = Node(self,0,False,0)
        self.activeNode = root 
        self.activeEdge = None 
        self.activeEdgeChar = None
        self.activeLength = 0
        self.end = -1
        self.root = root
        self.remaining = []
        
    def getActivePoint(self):
        # incorporate logic for jumping internal nodes
        if(self.activeLength == 0):
            return self.activeNode,None,0 ;
        if(self.activeEdgeChar is None):
            return self.activeNode,None,0
        previousLength = self.activeLength
        print("Tree.getActivePoint activeEdgeChar ",self.activeEdgeChar)
        activePoint,activeEdge,activeLength = self.activeNode.charToEdge[self.activeEdgeChar].getPoint(self.activeLength)
        #incorporate the logic for changing remaining[] in both if and else 
        if(isinstance(activePoint,Point)):
            print("Tree.getActivePoint activePoint is Point ",activePoint)
            self.activeNode = activePoint.node ;
            self.activeEdge = activePoint.edge ;
            self.activeLength = activePoint.distance ;
        else:
            print("Tree.getActivePoint activePoint is Node ",activePoint)
            self.activeNode = activePoint ;
            self.activeEdge = None ;
            self.activeLength = 0;
        lengthDifference = previousLength - self.activeLength ;
        self.remaining = self.remaining[lengthDifference:]
        print("Tree.getActivePoint after updation activeNode, ",self.activeNode," activeEdge ",self.activeEdge," activeLength ",self.activeLength," remaining ",self.remaining)
        return self.activeNode,self.activeEdge,self.activeLength
        
    def getWord(self):
        return self.word
        
    def __repr__(self):
        return "Tree {0},{1},{2},{3},{4}".format(self.word,self.activeNode,self.activeEdge,self.activeLength,self.end,self.remaining)
    
    def getEnd(self):
        return self.end
    
    def getNodeCount(self):
        return self.nodeCount;
    
    def getNext(self,char):
        if(self.activeLength == 0):
            print("Tree.getNext going for next of node ")
            return self.activeNode.getNext(char)
        print("Tree.getNext going for next of point")
        point = Point(self.activeNode,self.activeEdge,self.activeLength)
        return point.getNext(char)
        
class Edge():
    def __init__(self,start,end,fromNode,toNode,char):
        self.start = start ;
        self.end = end ;
        self.fromNode = fromNode ; 
        self.toNode = toNode 
        self.char = char
    
    def getPoint(self,activeLength):
        length = self.getLength()
        print("Edge.getPoint",self,length,activeLength)
        if(activeLength < length):
            print("Edge.getPoint returning point {0} distance away from {1} edged at {2}".format(activeLength,self.fromNode,self.char))
            return Point(self.fromNode,self,activeLength),self.char,activeLength
        elif(activeLength == length):
            print("Edge.getPoint returning ",self.toNode)
            return self.toNode,None,0
        else:
            word = self.getWord()
            nextChar = word[self.start + length];
            print("Edge.getPoint forwarding to {0}".format(self.toNode))
            return self.toNode.getPointFromChar(nextChar,activeLength - length)
        
    def getWord(self):
        return self.fromNode.getWord()
    
    def __repr__(self):
        return "Edge {0}->{1},{2}->{3}".format(self.start,self.getEnd(),self.fromNode,self.toNode)
    
    def getEnd(self):
        if(self.end is None):
            return self.fromNode.getEnd()
        return self.end
    
    def getLength(self):
        return self.getEnd() - self.start + 1
    # implement __hash__ and __eq__
        
class Node():
    def __init__(self,parentTree,nodeNumber,isLeaf,start):
        self.parentTree = parentTree
        self.charToEdge = dict()
        self.nodeNumber = nodeNumber ;
        self.isLeaf = isLeaf
        self.start = start
        parentTree.nodeCount += 1
        self.suffixLink = None
    
    def getWord(self):
        return self.parentTree.getWord();
    
    def getPointFromChar(self,char,length):
        if(length == 0):
            print("Node.getPointFromChar returning ",self)
            return self,None,0;
        edge = self.charToEdge.get(char);
        print("Node.getPointFromChar fowarding to ",edge)
        return edge.getPoint(length)
    
    def __repr__(self):
        return "Node {0}".format(str(self.nodeNumber));
    
    def createLeaf(self,char,start):
        leafNode = Node(self.parentTree,self.parentTree.getNodeCount(),True,start);
        edge = Edge(leafNode.start,None,self,leafNode,char)
        self.charToEdge[char] = edge
        
    def getNext(self,char):
        print("in getNextOf Node",str(self))
        charToEdge = self.charToEdge
        if(char in charToEdge):
            print("Node.getNext found {0} edge , forwarding to {1}".format(char,charToEdge[char]))
            return charToEdge[char].getPoint(1)[0]
        else:
            print("Node.getNext {0} edge not found , returning none ".format(char))
            return None
        
    def getEnd(self):
        return self.parentTree.getEnd()
        
        
class Point():
    def __init__(self,node,edge,distance):
        self.node = node ;
        self.edge = edge
        self.distance = distance ;
        
    def getWord(self):
        return self.node.getWord()
    
    def split(self,char,start):
        print("Point.split creating internal node ",self.node.parentTree.getNodeCount())
        internalNode = Node(self.node.parentTree,self.node.parentTree.getNodeCount(),False,self.edge.start)
        print("Point.split creating leaf node ",self.node.parentTree.getNodeCount())
        leafNode = Node(self.node.parentTree,self.node.parentTree.getNodeCount(),True,start);
        oldLeafNode = self.edge.toNode ;
        #logic to set the values for the below edge 
        internalToOldLeafEdge = Edge()
        edge = self.edge
        print("Point.split edge before ",edge)
        edge.toNode = internalNode ;
        edge.end = edge.start + self.distance - 1
        print("Point.split edge after ",edge)
        newEdge = Edge(self.node.getEnd(),None,internalNode,leafNode,char)
        print("Point.split new edge ",newEdge)
        word = self.getWord()
        char = word[edge.getEnd()];
        internalNode.charToEdge[char]=newEdge
        print("Point.split new edge added with {0} as {1}".format(char,newEdge))
        
    def getNext(self,char):
        #recheck the logic and incorporate for checking of char match before returning the result
        print("Point.getNext",str(self)," self.distance ",self.distance , " self.edge.length ",self.edge.getLength())
        if(self.distance < self.edge.getLength()):
            word = self.getWord()
            index = self.edge.start + self.distance
            foundChar = word[index]
            print("Point.getNext ",word,index,foundChar,char)
            if(char == foundChar):
                return Point(self.node,self.edge,self.distance+1)
            return None
        elif(self.distance == self.edge.getLength()):
            return self.edge.toNode.getNext(char);
        else:
            raise('should not reach here')
    def __repr__(self):
        return "Point {0} -> {1} at {2}".format(self.node,self.edge,self.distance);

def generateSuffixTree(word):
    nChars = len(word)
    tree = SuffixTree(word)
    # phase
    for i in range(nChars):
        tree.end = i
        char = word[i]
        print("-- i=",i," ",char)
        print("-- tree ",tree)
        continueOuter = False;
        point,activeEdge,activeLength = tree.getActivePoint()
        nextPoint = tree.getNext(char)
        #rule 1
        print("-- end is now at ", i)
        print("-- nextPoint is ",nextPoint)
        tailNode = None;
        if(nextPoint is None):
            #rule 2
            if(activeLength == 0):
                #create leaf
                    tree.activeNode.createLeaf(char,i)
            else:
                #split to create internal and leaf
                point = Point(tree.activeNode,tree.activeEdge,tree.activeLength)
                point.split(char,i)
            tree.remaining.append(char)
            print("-- after insertion , remaining ",tree.remaining)
            if(tree.activeNode == tree.root):
                tree.remaining = tree.remaining[1:]
                print("-- activeNode is root ")
                if(len(tree.remaining)>0):
                    tree.activeEdgeChar = tree.remaining[0]
                    print("-- tree.activeEdge ",tree.activeEdgeChar)
                else:
                    tree.activeEdge = None;
                    print("-- tree.activeEdge ",tree.activeEdgeChar)
                tree.activeLength = max(0,tree.activeLength - 1)
                print("-- tree.activeLength ",tree.activeLength)
            else:
                if(tree.activeNode.suffixLink is None):
                    print("-- no suffix link ")
                    tree.activeNode = tree.root
                    print("-- tree.activeNode ",tree.activeNode)
                else:
                    print("-- suffix link found")
                    tree.activeNode = tree.activeNode.suffixLink
                    print("-- tree.activeNode ",tree.activeNode)
        else:
            # rule 3
            # this means you found the next char on an edge 
            print("-- nextPoint is on an edge ",nextPoint,tree.remaining)
            if(tree.activeEdge is None):
                tree.activeEdgeChar = char
            tree.activeLength +=1 ;
            tree.remaining.append(char)
            continueOuter = True ;
            print("-- rule 3 , should breakout ")
            print("-- tree.activeLength ",tree.activeLength)
            print("-- remaining ",tree.remaining)
        if(continueOuter):
            print("-- rule 3 has set continue outer to true")
            continue;
        while len(tree.remaining)>0:
            print("-- tackle remaining ",tree.remaining)
            print("-- i2=",i," ",char)
            print("-- tree2 ",tree)
            point,activeEdge,activeLength = tree.getActivePoint()
            nextPoint = tree.getNext(char)
            print("-- point ",point)
            print("-- nextPoint ",nextPoint)
            if(nextPoint is None):
                #rule 2
                print("-- nextPoint is None")
                #it means we reached an internal node and there is nowhere else to go , create a leaf node 
                if(activeLength == 0):
                    #create leaf
                    tree.activeNode.createLeaf(char,i)
                else:
                    #split to create internal and leaf
                    point = Point(tree.activeNode,tree.activeEdge,tree.activeLength)
                    point.split(char,i)
                print("-- remaining ",tree.remaining)
                if(tree.activeNode == tree.root):
                    tree.remaining = tree.remaining[1:]
                    if(len(tree.remaining)>0):
                        tree.activeEdgeChar = tree.remaining[0]
                        print("-- activeEdgeChar ",tree.activeEdgeChar)
                    else:
                        tree.activeEdgeChar = None;
                        tree.activeLength = max(0,tree.activeLength - 1)
                        print("-- nothing remains , activeEdgeChar ",tree.activeEdgeChar," activeLength ",tree.activeLength)
                else:
                    print("-- active node is not root")
                    if(tree.activeNode.suffixLink is None):
                        tree.activeNode = tree.root
                        print("-- no suffix link, activeNode ",tree.activeNode)
                    else:
                        tree.activeNode = tree.activeNode.suffixLink
                        print("-- suffix link found ,active node ",tree.activeNode)
            else:
                # rule 3
                # this means you found the next char on an edge 
                print("-- nextPoint is on an edge ",nextPoint,tree.remaining)
                if(tree.activeEdge is None):
                    tree.activeEdgeChar = char
                tree.activeLength +=1 ;
                print("-- remaining ",tree.remaining)
                break;
    return tree
            
        
def main():
    
    print(generateSuffixTree('mississippi$'))
    
main()
```
